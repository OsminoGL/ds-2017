```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
```

```python
df = pd.read_csv("data/mpg.csv")
```

```python
df.head(5)
```

```python
class_list = df['class'].tolist()
```

```python
markers = ['v','o','>','^','<','.','s']

size = [x*int(round(df.cty.mean())) for x in range(1,8)]
```

```python
int(round(df.cty.mean()))
```

```python
g = sns.lmplot('cty', 'displ',
           data=df,
           fit_reg=False,
           palette='Set1',
           markers = markers,
           hue="class",
           scatter_kws={"s": size})
g.fig.set_size_inches(18, 6)
None
```

```python
plt.figure(figsize=(12,8))
plot = sns.swarmplot(df.cty, df.displ, hue=class_list, size=7, palette='Set1')
None
```
