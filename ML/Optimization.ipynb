```python
import math as m
```

```python
input = [x for x in range(-5,6)]

output = []
```

```python
# Optimize function
# F(x) = xsin(x) => d=sin(x)xcos(x)

for x in input:
    output.append(m.sin(x)*x*m.cos(x))
print(min(output))
```

```python
# ploting the interval 

%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np

def graph(formula, x_range):  
        x = np.array(x_range)  
        y = eval(formula)
        plt.plot(x, y)  
        plt.show()
```

```python
graph('np.sin(x)*x*np.cos(x)', range(-5,6))
```

```python
graph('x*np.sin(x)', range(-5,6))
```

```python
# F(x) = x/cos(x) => d=(xtan(x)+1)/sec(x)
output = []

for x in input:
    output.append((m.tan(x)+1)/(1/m.cos(x)))
print(min(output))
```

```python
graph('(np.tan(x)+1)/(1/np.cos(x))', range(-5,6))
```

```python
# F(x) = exp(x)*sqrt(x) => d=(eˆx*(2x+1))/2*sqrt(x)
output = []

for x in input:
    if x > 0:
        output.append(((m.exp(1)**x)*(2*x+1))/(2*m.sqrt(x)))
print(min(output))
```

```python
graph('((np.exp(1)**x)*(2*x+1))/(2*np.sqrt(x))', range(1,6))
```

```python
inputX = [x for x in range(-5,6)]
inputY = [y for y in range(-10,11)]
```

```python
output = []

for y in inputY:
    for x in inputX:
        output.append(y*m.cos(x))
print(min(output))
```

```python
output = []

for y in inputY:
    for x in inputX:
        output.append(y*m.tan(x)*(1/m.cos(x)))
print(min(output))
```

```python
output = []

for y in inputY:
    for x in inputX:
        if x > 0:
            output.append((m.exp(1)**-y)/2*sqrt(x))
print(min(output))
```
